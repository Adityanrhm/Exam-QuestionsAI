```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats


# laod dataset
file_path_dataset = ('2019.csv')
df_WorldHappiness = pd.read_csv(file_path_dataset)

# Preview Dataset
df_WorldHappiness
```

```python
# ANSWER NUMBER 1 World Happiness
```

```python
df_WorldHappiness = df_WorldHappiness.drop(columns=['Overall rank'])
numeric_columns = df_WorldHappiness.select_dtypes(include=[np.number]).columns

mean_values = df_WorldHappiness[numeric_columns].mean()
median_values = df_WorldHappiness[numeric_columns].median()
modus_values = df_WorldHappiness[numeric_columns].mode().iloc[0]
range_values = df_WorldHappiness[numeric_columns].max() - df_WorldHappiness[numeric_columns].min()

q1_values = df_WorldHappiness[numeric_columns].quantile(0.25)
q3_values = df_WorldHappiness[numeric_columns].quantile(0.75)
iqr_values = q3_values - q1_values

standard_deviasi = df_WorldHappiness[numeric_columns].std()
variance = df_WorldHappiness[numeric_columns].var()
```

```python
statistics_summary = pd.DataFrame({
    "Mean": mean_values,
    "Median": median_values,
    "Modus": modus_values,
    "Range Values": range_values,
    "Q1": q1_values,
    "Q3": q3_values,
    "IQR": iqr_values,
    "Standard Deviasi": standard_deviasi,
    "Variance": variance
})

statistics_summary.T
```

```python
z_scores = df_WorldHappiness[numeric_columns].apply(stats.zscore)
z_scores.head()
```

```python
# ANSWER NUMBER 2 World Happiness 
```

```python
lower_bound = q1_values - 1.5 * iqr_values
upper_bound = q3_values + 1.5 * iqr_values

iqr_outliers = (df_WorldHappiness[numeric_columns] < lower_bound) | (df_WorldHappiness[numeric_columns] > upper_bound)
outliers_iqr = iqr_outliers.sum(axis=0)

zscore_outliers = abs(z_scores[numeric_columns]) > 3
outliers_zscore = zscore_outliers.sum(axis=0)

df = {
    "Outliers (IQR)": outliers_iqr,
    "Outliers (Z-score)": outliers_zscore
}

outliers_df = pd.DataFrame(df)
print(outliers_df)
```

```python
# ANSWER NUMBER 3 World Happiness
```

```python
plt.figure(figsize=(15, 10))

for i, column in enumerate(numeric_columns):
    plt.subplot(len(numeric_columns), 2, 2*i+1)
    sns.boxplot(x=df_WorldHappiness[column])
    plt.title(f"Bloxplot of {column}")

    plt.subplot(len(numeric_columns),2, 2*i+2)
    sns.histplot(df_WorldHappiness[column], kde=True, bins=20)
    plt.title(f"Histogram of {column}")
    
plt.tight_layout()
plt.show()
```

```python
#ANSWER NUMBER 4 WORLD HAPPINESS
```

```python
df_numeric = df_WorldHappiness.select_dtypes(include=[np.number])

covariance = df_numeric.cov()["Score"].drop("Score")
pearson_corr = df_numeric.corr(method="pearson")["Score"].drop("Score")
spearman_corr = df_numeric.corr(method="spearman")["Score"].drop("Score")
kendall_corr = df_numeric.corr(method="kendall")["Score"].drop("Score")

summary_correlation = pd.DataFrame({
    "Covariance": covariance,
    "Pearson Correlation": pearson_corr,
    "Spearman Correlation": spearman_corr,
    "Kendall Correlation": kendall_corr
})

# df_correlation = pd.DataFrame(summary_correlation).T
summary_correlation.T
```

```python
top_tgp = df_WorldHappiness.nlargest(10, "GDP per capita")[[ "Country or region","GDP per capita"]]
top_health = df_WorldHappiness.nlargest(10, "Healthy life expectancy")[["Country or region","Healthy life expectancy" ]]
top_corruptions = df_WorldHappiness.nlargest(10, "Perceptions of corruption")[["Country or region", "Perceptions of corruption"]]


fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.barplot(y=top_tgp["Country or region"], x=top_tgp["GDP per capita"], ax=axes[0], palette="viridis")
axes[0].set_title("Top 10 Country by GDP per capita")
axes[0].set_xlabel("GDP per capita")
axes[0].set_ylabel("Country")

sns.barplot(y=top_health["Country or region"], x=top_health["Healthy life expectancy"], ax=axes[1], palette="coolwarm")
axes[1].set_title("Top 10 Country by Healthy life expectancy")
axes[1].set_xlabel("Healthy life expectancya")
axes[1].set_ylabel("Country")

sns.barplot(y=top_corruptions["Country or region"], x=top_corruptions["Perceptions of corruption"], ax=axes[2], palette="magma")
axes[2].set_title("Top 10 Country by Perceptions of corruption")
axes[2].set_xlabel("Perceptions of corruption")
axes[2].set_ylabel("Country")

sns.barplot()

plt.tight_layout()
plt.show()
```
