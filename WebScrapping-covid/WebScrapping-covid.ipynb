```python
import pandas as pd
from bs4 import BeautifulSoup as bs4
import requests
import folium

url_raw_gps_indonesia = "https://raw.githubusercontent.com/LintangWisesa/Indonesia-Covid19-Maps/master/data/gps_indonesia.json"
url_raw_statistics_covid = 'https://en.wikipedia.org/wiki/Statistics_of_the_COVID-19_pandemic_in_Indonesia'
```

```python
response_gps_indonesia = requests.get(url_raw_gps_indonesia)
req_statistics_covid = requests.get(url_raw_statistics_covid).text

radius_covid
```

```python

statistics_covid_soup = bs4(req_statistics_covid, 'lxml')
radius_covid = statistics_covid_soup.find_all('table')

pd.set_option('display.max_colwidth', None)
df_gps_indonesia = pd.read_json(url_raw_gps_indonesia)
df_gps_indonesia.index = df_gps_indonesia.index + 1

rows = radius_covid[0].find_all('tr')
data = []

# Extract table data
for row in rows:
    cols = row.find_all('td')
    cols = [jawa.text.strip() for jawa in cols]
    data.append(cols)

columns = ['Cases', 'Recoveries', 'Deaths', 'Active', 'Cases per 100,000 population', 
           'Recovery Rate', 'Fatality Rate', 'Official Website']

provinces = ['Aceh', 'Bali', 'Bangka Belitung', 'Banten', 'Bengkulu', 'Central Java', 'Central Kalimantan', 
             'Central Sulawesi', 'East Java', 'East Kalimantan', 'East Nusa Tenggara', 'Gorontalo', 'Jakarta', 
             'Jambi', 'Lampung', 'Maluku', 'North Maluku', 'North Kalimantan', 'North Sulawesi', 'North Sumatera', 
             'Papua', 'Riau', 'Riau Islands', 'South Kalimantan', 'South Sulawesi', 'Southeast Sulawesi', 
             'South Sumatera', 'West Java', 'West Kalimantan', 'West Nusa Tenggara', 'West Papua', 'West Sulawesi', 
             'West Sumatra', 'Yogyakarta']

df_statistics_covid = pd.DataFrame(data, columns=columns)


unwanted_words = [
    "None"
]

df_statistics_covid = df_statistics_covid[
    ~df_statistics_covid.apply(lambda row: any(word in row.to_string() for word in unwanted_words), axis=1)
]

df_statistics_covid.insert(0, 'Provinsi', provinces)
```

```python
map_indonesia = folium.Map(location=[-2.325613, 118.005351],zoom_start=5)

merge_df = pd.merge(df_gps_indonesia, df_statistics_covid, on="Provinsi", how="inner")

for _, rows in merge_df.iterrows():
    prov = rows["Provinsi"] 
    lat = rows["latitude"]
    longt = rows["longitude"]
    icon_logo = rows["logo"]
    confirmed = rows["Active"] 
    recovered = rows["Recoveries"] 
    deaths = rows["Deaths"] 

    
    folium_icon = folium.CustomIcon(
        icon_logo,
        icon_size=(28, 32)
    )

    popup_html = f"""
    <div style="font-family: Arial; font-size: 14px; background-color: #f0f0f0; padding: 10px; border-radius: 8px; width: 200px; border: 1px solid #ccc;">
        <h4 style="font-size: 20px; font-weight: bold; color: #333;">{prov}</h4>
        <div style="display: flex; flex-direction: column; gap: 5px;">
            <div style="display: flex; justify-content: space-between;">
                <b style="color: red;">Confirmed:</b><span style="color: red;">{confirmed}</span>
            </div>
            <div style="display: flex; justify-content: space-between;">
                <b style="color: green;">Recovered:</b><span style="color: green;">{recovered}</span>
            </div>
            <div style="display: flex; justify-content: space-between;">
                <b style="color: black;">Deaths:</b><span style="color: black;">{deaths}</span>
            </div>
        </div>
    </div>
    """
    
    
    folium_marker = folium.Marker(
        location=[lat, longt],
        icon=folium_icon,
        tooltip=prov,
        popup=popup_html
    ).add_to(map_indonesia)
map_indonesia
```
